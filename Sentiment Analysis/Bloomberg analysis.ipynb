```python
import pandas as pd

df = pd.read_csv('../data/News Articles/processed/Bloomberg_with_scores.csv', encoding='ISO-8859-1', index_col='Date')
df.index = pd.to_datetime(df.index)
df = df.sort_index()
```

```python
df.columns
```

```python
import datetime as dt

cols = ['distilbert_score', 'svc_score', 'gbdt_score',
       'knn_score', 'vader_score']
df1 = df.copy()[cols]

df1 = df1.groupby(df1.index.date).mean()
print(df1)
```

```python
df_daily = pd.read_csv('../data/Stock Data/SPY_2006_2013_daily.csv', index_col='Date')
df_daily.index = pd.to_datetime(df_daily.index)
```

```python
df1['Close'] = df1.index.map(df_daily['Close'])

df1['3d'] = (df1.index + pd.Timedelta(days=3)).map(df_daily['Close']) - (df1.index + pd.Timedelta(days=4)).map(df_daily['Close'])
df1['9d'] = (df1.index + pd.Timedelta(days=9)).map(df_daily['Close']) - (df1.index + pd.Timedelta(days=12)).map(df_daily['Close'])
df1['15d'] = (df1.index + pd.Timedelta(days=15)).map(df_daily['Close']) - (df1.index + pd.Timedelta(days=20)).map(df_daily['Close'])
df1['1m'] = (df1.index + pd.Timedelta(days=30)).map(df_daily['Close']) - (df1.index + pd.Timedelta(days=40)).map(df_daily['Close'])
df1['2m'] = (df1.index + pd.Timedelta(days=60)).map(df_daily['Close']) - (df1.index + pd.Timedelta(days=80)).map(df_daily['Close'])
df1['3m'] = (df1.index + pd.Timedelta(days=90)).map(df_daily['Close']) - (df1.index + pd.Timedelta(days=120)).map(df_daily['Close'])

df1
```

```python
df1.columns
```

```python
# List of topic columns (manually extracted)
topic_cols = [
    'cluster', 'Climate & Environment', 'Consumer & Retail', 'Corporate Sector',
    'Crypto & Digital Assets', 'Education & Knowledge', 'Energy & Commodities',
    'Financial Markets', 'Geopolitics & Policy', 'Health & Biotech',
    'Legal & Regulation', 'Macro Finance', 'Real Estate & Housing',
    'Society & Labor', 'Tech & Innovation', 'Transport & Logistics', 'Uncertain'
]

sentiment_cols = ['distilbert_score', 'svc_score',
       'gbdt_score', 'knn_score', 'vader_score']

# Close price column
close_col = ['Close']

# Time horizon columns
horizon_cols = ['3d', '9d', '15d', '1m', '2m', '3m']

# Dictionary to store each mini-DataFrame
mini_dfs = {}

for h in horizon_cols:

    # Build mini_df without including `h` or `close_col` directly
    mini_df = df1[sentiment_cols + [h]].copy()
    mini_df['price_diff'] = df1[h].copy()

    mini_dfs[h] = mini_df
```

```python
import statsmodels.api as sm
from scipy.stats import pearsonr
import pandas as pd

# Store output
summary_stats = []

for horizon, mini_df in mini_dfs.items():
    df_reg = mini_df.dropna()

    X = df_reg[sentiment_cols]
    y = df_reg['price_diff']

    X = sm.add_constant(X)

    model = sm.OLS(y, X).fit()
    y_pred = model.fittedvalues
    corr, _ = pearsonr(y, y_pred)
    
    for col in sentiment_cols:
        summary_stats.append({
            'horizon': horizon,
            'r_squared': model.rsquared,
            'correlation': corr,
            'variable': col,
            'coef': model.params[col],
            't_stat': model.tvalues[col],
            'p_value': model.pvalues[col],
        })

summary_df = pd.DataFrame(summary_stats)

# Save to CSV
summary_df.to_csv('../data\Results Charts Models/trump_linear_regression_summary_3to1.csv', index=False)

# Also print for quick check
print(summary_df.round(4))
```
