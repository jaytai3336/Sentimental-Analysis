```python
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    if not isinstance(text, str) or text is None:
        return ""
    tokens = word_tokenize(text.lower())  # Tokenize and lowercase
    cleaned_tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha() and token not in stop_words]
    test_sentence = " ".join(cleaned_tokens)
    return test_sentence

df_grok['processed_text'] = df_grok['Headline'].apply(preprocess)
df_grok['processed_text'].head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import LinearSVC

pipeline_svc = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('model', LinearSVC())
])

pipeline_gb = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('model', GradientBoostingClassifier())
])

pipeline_knn = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('model', KNeighborsClassifier())
])

X = df_grok['processed_text']
y = df_grok['Sentiment_label']
pipeline_svc.fit(X,y)
pipeline_gb.fit(X,y)
pipeline_knn.fit(X,y)
```

```python
import pandas as pd

df = pd.read_csv('../data/News Articles/processed/trump_processed_tagged.csv', encoding='ISO-8859-1')
df = df.drop(columns=['matched_keywords'], axis=1)
df = df.dropna()
df
```

```python
df['sentiment'] = df['sentiment'].apply(lambda x: 1 if x == 'POSITIVE' else -1)
df['distilbert_score'] = df['sentiment_score'] * df['sentiment']
df.drop(columns=['sentiment_score', 'sentiment'], axis=1, inplace=True)
```

```python
import numpy as np 
x = df['processed_text']
y1 = pipeline_svc.predict(x)
y1 = [1 if i == 'POS' else (-1 if i == 'NEG' else 0) for i in y1 ]
df['svc_score'] = y1

y2 = pipeline_gb.predict(x)
y2 = [1 if i == 'POS' else (-1 if i == 'NEG' else 0) for i in y2]
df['gbdt_score'] = y2

y3 = pipeline_knn.predict(x)
y3 = [1 if i == 'POS' else (-1 if i == 'NEG' else 0) for i in y3 ]
df['knn_score'] = y3
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Define thresholds for labeling
threshold_upper = 0
threshold_lower = -0.02

# Function to compute aggregate score
def vader_aggregate_score(sentence):
    compound = analyzer.polarity_scores(sentence)['compound']
    
    # Assign label
    if compound >= threshold_upper:
        label = 1  # POS
    elif compound <= threshold_lower:
        label = -1  # NEG
    else:
        label = 0  # NEU
    
    # Combine label and strength
    return label * abs(compound)

# Apply to DataFrame
df['vader_score'] = df['processed_text'].astype(str).apply(vader_aggregate_score)
```

```python
from flair.data import Sentence
from flair.nn import Classifier
import pandas as pd

# Load Flair sentiment model
tagger = Classifier.load('sentiment')

# Function to compute aggregate Flair score
def get_flair_agg_score(text):
    sentence = Sentence(str(text))  # Ensure it's a string
    tagger.predict(sentence)
    label = sentence.labels[0].value        # 'POSITIVE' or 'NEGATIVE'
    confidence = sentence.labels[0].score   # Between 0.5 and 1.0
    return confidence if label == 'POSITIVE' else -confidence

# Apply to your dataset
df['flair_score'] = df['processed_text'].apply(get_flair_agg_score)
```

```python
import pandas as pd
df = pd.read_csv('../data/News Articles/processed/trump_processed_tagged_scores.csv', encoding='ISO-8859-1')
df
```

```python
# List of topic columns (manually extracted)
topic_cols = [
    'cluster', 'Climate & Environment', 'Consumer & Retail', 'Corporate Sector',
    'Crypto & Digital Assets', 'Education & Knowledge', 'Energy & Commodities',
    'Financial Markets', 'Geopolitics & Policy', 'Health & Biotech',
    'Legal & Regulation', 'Macro Finance', 'Real Estate & Housing',
    'Society & Labor', 'Tech & Innovation', 'Transport & Logistics', 'Uncertain'
]

sentiment_cols = ['distilbert_score', 'svc_score',
       'gbdt_score', 'knn_score', 'vader_score', 'flair_score']

# Close price column
close_col = ['Close']

# Time horizon columns
horizon_cols = ['1h', '2h', '5h', '12h', '24h']
horizon_lag_cols = ['1h20min', '2h40min', '5h100min', '12h240min', '24h480min']

# Dictionary to store each mini-DataFrame
mini_dfs = {}

for h in range(0,len(horizon_cols)):
    
#     change to range

    # Compute price difference
    price_diff = df[horizon_cols[h]] - df[horizon_lag_cols[h]]

    # Build mini_df without including `h` or `close_col` directly
    mini_df = df[topic_cols + sentiment_cols].copy()
    mini_df['price_diff'] = price_diff

    mini_dfs[horizon_cols[h]] = mini_df
```

```python
import statsmodels.api as sm
from scipy.stats import pearsonr
import pandas as pd

# All 6 sentiment features
sentiment_cols = ['distilbert_score', 'svc_score',
       'gbdt_score', 'knn_score', 'vader_score', 'flair_score']

# Store output
summary_stats = []

for horizon, mini_df in mini_dfs.items():
    df_reg = mini_df.dropna()

    X = df_reg[sentiment_cols]
    y = df_reg['price_diff']

    X = sm.add_constant(X)

    model = sm.OLS(y, X).fit()
    y_pred = model.fittedvalues
    corr, _ = pearsonr(y, y_pred)
    
    for col in sentiment_cols:
        summary_stats.append({
            'horizon': horizon,
            'r_squared': model.rsquared,
            'correlation': corr,
            'variable': col,
            'coef': model.params[col],
            't_stat': model.tvalues[col],
            'p_value': model.pvalues[col],
        })

summary_df = pd.DataFrame(summary_stats)

# Save to CSV
summary_df.to_csv('../data\Results Charts Models/trump_linear_regression_summary_3to1.csv', index=False)

# Also print for quick check
print(summary_df.round(4))
```

```python
import statsmodels.api as sm
import pandas as pd

# Sentiment predictors
sentiment_cols = ['distilbert_score', 'svc_score', 'gbdt_score', 'knn_score', 'flair_score', 'vader_score']

results = []

for horizon, mini_df in mini_dfs.items():
    df_reg = mini_df.dropna()

    # Create binary direction target: 1 if price_diff > 0 else 0
    df_reg['direction'] = (df_reg['price_diff'] > 0).astype(int)

    X = df_reg[sentiment_cols]
    y = df_reg['direction']

    X = sm.add_constant(X)

    try:
        model = sm.Logit(y, X).fit(disp=False)  # disp=False hides fit output
    except Exception as e:
        print(f"Logistic regression failed for {horizon}: {e}")
        continue

    for col in sentiment_cols:
        results.append({
            'horizon': horizon,
            'variable': col,
            'coef': model.params[col],
            't_stat': model.tvalues[col],
            'p_value': model.pvalues[col],
            'model_llf': model.llf,          # log-likelihood of model
            'model_pseudo_r2': model.prsquared  # McFadden's pseudo R²
        })

# Create summary DataFrame
summary_df = pd.DataFrame(results)

# Pivot table for easier reading (optional)
pivot_table = summary_df.pivot(index='horizon', columns='variable', values=['coef', 't_stat', 'p_value'])

# Save to CSV
summary_df.to_csv('logistic_regression_sentiment_direction.csv', index=False)
# summary_df.to_csv('../data\Results Charts Models/trump_log_regression_summary.csv', index=False)

print(pivot_table.round(4))
```
