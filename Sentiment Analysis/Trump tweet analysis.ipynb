```python
import pandas as pd
```

```python
df = pd.read_csv('../data/News Articles/processed/trump_analysed.csv', encoding='ISO-8859-1')
df['datetime'] = pd.to_datetime(df['datetime'], format='%B %d, %Y @ %I:%M %p ET')
df.set_index('datetime', inplace=True)
df = df.drop(['matched_keywords'], axis=1)
df.dropna()
df.head()
```

```python
# drop retweets and drop those with confidence score below .997

df_cleaned = df[df['isRetweet'] == False]
df_cleaned = df[df['sentiment_score']>=0.997]
df_cleaned.describe()
```

```python
df_cleaned.to_csv('../data/News Articles/processed/trump_filtered.csv')
```
