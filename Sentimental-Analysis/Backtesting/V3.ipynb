```python
class Vader(bt.Strategy):
    params = (
        ('size', 1),  # Number of units to buy/sell
        ('hold_time_mins', 30),  # Minutes to hold position
        ('commission', 0.0003),  # 0.03% commission
    )

    def __init__(self):
        self.order_list = []
        self.order = None
        self.trade_history = []
        self.cumulative_pnl = 0
        self.trade_count = 0
        self.unrealized_pnl = 0
        self.current_position = 0  # Track current position size
        self.total_commission = 0
        self.initial_value = self.broker.getvalue()

    def UpdateUnrealizedPnl(self, date):
        """Calculate unrealized P&L for open positions"""
        current_price = self.datas[0].close[0]
        self.unrealized_pnl = 0

        for order_info in self.order_list:
            if order_info['direction'] == 'BUY':
                self.unrealized_pnl += (current_price - order_info['price']) * order_info['size']
            else:
                self.unrealized_pnl += (order_info['price'] - current_price) * order_info['size']

    def ExecuteTrade(self, direction, date):
        """Execute trade and log entry"""
        price = self.datas[0].close[0]
        size = self.params.size

        if direction == 'BUY':
            self.order = self.buy(size=size)
            self.current_position += size
            print(f'{date} - ENTERED LONG {size} @ {price:.2f}')
        else:
            self.order = self.sell(size=size)
            self.current_position -= size
            print(f'{date} - ENTERED SHORT {size} @ {price:.2f}')

        self.order_list.append({
            'order': self.order,
            'order_date': date,
            'close_date': date + pd.Timedelta(minutes=self.params.hold_time_mins),
            'price': price,
            'size': size,
            'direction': direction,
            'trade_id': self.trade_count
        })
        self.trade_count += 1

    def CheckOrderExpiry(self, date):
        """Close expired positions and calculate P&L"""
        for order_info in self.order_list[:]:  # Iterate over copy
            if date >= order_info['close_date']:
                exit_price = self.datas[0].close[0]
                size = order_info['size']
                trade_id = order_info['trade_id']

                if order_info['direction'] == 'BUY':
                    self.order = self.sell(size=size)
                    self.current_position -= size
                    pnl = (exit_price - order_info['price']) * size
                else:
                    self.order = self.buy(size=size)
                    self.current_position += size
                    pnl = (order_info['price'] - exit_price) * size

                self.cumulative_pnl += pnl

                # Print trade exit
                print(f'{date} - CLOSED {"LONG" if order_info["direction"] == "BUY" else "SHORT"} '
                      f'trade #{trade_id} @ {exit_price:.2f} | '
                      f'P&L: {pnl:.2f} | '
                      f'Total P&L: {self.cumulative_pnl:.2f}')

                # Record trade
                self.trade_history.append({
                    'id': trade_id,
                    'direction': order_info['direction'],
                    'entry': order_info['price'],
                    'exit': exit_price,
                    'entry_time': order_info['order_date'],
                    'exit_time': date,
                    'pnl': pnl
                })

                self.order_list.remove(order_info)  # Remove from active orders

    def notify_order(self, order):
        """Handle order notifications"""
        if order.status in [order.Completed]:
            commission = order.executed.price * abs(order.executed.size) * self.p.commission
            self.total_commission += commission
            if order.isbuy():
                pass  # Buy order completed
            elif order.issell():
                pass  # Sell order completed

    def stop(self):
        """Final cleanup and reporting"""
        # Close all remaining positions
        self.CloseAllPositions()

        # Final performance report
        print('\n=== FINAL STRATEGY PERFORMANCE ===')
        print(f'Total Trades Executed: {self.trade_count}')
        print(f'Final Cumulative P&L: {self.cumulative_pnl:.2f}')

        if self.trade_history:
            winning_trades = sum(1 for t in self.trade_history if t['pnl'] > 0)
            win_rate = (winning_trades / len(self.trade_history)) * 100
            avg_win = sum(t['pnl'] for t in self.trade_history if t['pnl'] > 0) / max(1, winning_trades)
            avg_loss = sum(t['pnl'] for t in self.trade_history if t['pnl'] < 0) / max(1, (len(self.trade_history) - winning_trades))

            print(f'\n--- Performance Metrics ---')
            print(f'Win Rate: {win_rate:.1f}%')
            print(f'Average Winning Trade: {avg_win:.2f}')
            print(f'Average Losing Trade: {avg_loss:.2f}')
            print(f'Profit Factor: {self.CalculateProfitFactor():.2f}')

            # Print trade log
            print('\n--- Trade Log ---')
            for trade in self.trade_history:
                print(f"Trade #{trade['id']}: {trade['direction']} | "
                      f"Entry: {trade['entry']:.2f} | Exit: {trade['exit']:.2f} | "
                      f"P&L: {trade['pnl']:.2f}")

        final_value = self.broker.getvalue()
        print(f"\n=== COST BREAKDOWN ===")
        print(f"Initial Portfolio Value: {self.initial_value:.2f}")
        print(f"Trade P&L: {self.cumulative_pnl:.2f}")
        print(f"Total Commissions: -{self.total_commission:.2f}")
        print(f"Final Portfolio Value: {final_value:.2f}")
        a  = self.total_commission - self.cumulative_pnl + final_value
        print(f"Reconcilation: {a:.2f}")

    def CloseAllPositions(self):
        """Force close all open positions at market price"""
        if self.current_position != 0:
            current_price = self.datas[0].close[0]
            close_size = abs(self.current_position)

            if self.current_position > 0:  # Long position to close
                self.order = self.sell(size=close_size)
                pnl = (current_price - self.GetAverageEntryPrice()) * close_size
                print(f'\nCLOSING FINAL LONG POSITION: {close_size} @ {current_price:.2f}')
            else:  # Short position to close
                self.order = self.buy(size=close_size)
                pnl = (self.GetAverageEntryPrice() - current_price) * close_size
                print(f'\nCLOSING FINAL SHORT POSITION: {close_size} @ {current_price:.2f}')

            self.cumulative_pnl += pnl
            print(f'Final Position P&L: {pnl:.2f}')
            print(f'Total Cumulative P&L After Final Close: {self.cumulative_pnl:.2f}')
            self.current_position = 0

    def GetAverageEntryPrice(self):
        """Calculate average entry price for current position"""
        if not self.order_list:
            return 0

        total_value = 0
        total_size = 0

        for order in self.order_list:
            if order['direction'] == 'BUY':
                total_value += order['price'] * order['size']
                total_size += order['size']
            else:
                total_value -= order['price'] * order['size']
                total_size -= order['size']

        return total_value / total_size if total_size != 0 else 0

    def CalculateProfitFactor(self):
        """Calculate profit factor (gross profits / gross losses)"""
        gross_profit = sum(t['pnl'] for t in self.trade_history if t['pnl'] > 0)
        gross_loss = abs(sum(t['pnl'] for t in self.trade_history if t['pnl'] < 0))
        return gross_profit / gross_loss if gross_loss != 0 else float('inf')

    def next(self):
        PriceDate = self.datas[0].datetime.datetime(0)
        SentimentDate = self.datas[1].datetime.datetime(0)
        VaderTS = self.datas[1].vaderTS[0]
        VaderDS = self.datas[1].vaderDS[0]

        # First close any expired orders
        self.CheckOrderExpiry(PriceDate)

        # Calculate current unrealized P&L
        self.UpdateUnrealizedPnl(PriceDate)

        # Enter new trades based purely on Vader signals
        if PriceDate == SentimentDate:
            if VaderDS > 0 and VaderTS > 0:  # Sell signal
                self.ExecuteTrade('BUY', PriceDate)
                pass
            else:  # Buy signal
                #self.ExecuteTrade('SELL', PriceDate)
                pass

run(Vader, file_path='S1/Vader_plot.png')
```
